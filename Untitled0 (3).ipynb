```python
import tensorflow as tf
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt
import numpy as np

# 1. Load the MNIST dataset
mnist = tf.keras.datasets.mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# 2. Data Preprocessing
# Scale pixel values to 0-1
x_train, x_test = x_train / 255.0, x_test / 255.0

# CNN requires an extra dimension for the color channel (1 for grayscale)
x_train = x_train.reshape(-1, 28, 28, 1)
x_test = x_test.reshape(-1, 28, 28, 1)

# 3. Build the CNN Architecture
model = models.Sequential([
    # First Convolutional Layer to detect basic shapes
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    layers.MaxPooling2D((2, 2)),

    # Second Convolutional Layer for complex patterns
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),

    # Flatten and Dense layers for classification
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.2),
    layers.Dense(10, activation='softmax')
])

# 4. Compile the Model
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# 5. Train the Model
# We increase epochs to 10 for better accuracy
print("--- Training Started ---")
model.fit(x_train, y_train, epochs=10)

# 6. Evaluate the Model
print("\n--- Evaluation on Test Data ---")
test_loss, test_acc = model.evaluate(x_test, y_test, verbose=2)
print(f'Test Accuracy: {test_acc * 100:.2f}%')

# 7. Save Weights for your Streamlit App
# This is the file you need to download and put in your VS Code folder
model.save_weights('cnn_digit_weights.weights.h5')

# Trigger download automatically in Colab
from google.colab import files
files.download('cnn_digit_weights.weights.h5')

# 8. Quick Test Prediction
image_index = 0
plt.imshow(x_test[image_index].reshape(28, 28), cmap='gray')
plt.show()

prediction = model.predict(x_test[image_index].reshape(1, 28, 28, 1))
print(f"Model's Prediction: {np.argmax(prediction)}")
```
